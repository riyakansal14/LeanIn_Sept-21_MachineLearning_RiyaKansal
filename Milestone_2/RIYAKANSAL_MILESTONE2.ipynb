```python
import sqlite3
#creating database
conn = sqlite3.connect('spending_analysis.sqlite')
cur = conn.cursor()
```

```python
#Creating table in database
cur.execute('DROP TABLE IF EXISTS SPENDING_ANALYSIS')
cur.execute('''CREATE TABLE SPENDING_ANALYSIS (
CUSTOMERID INTEGER PRIMARY KEY,
GENRE TEXT,
AGE INTEGER,
Annual_Income INTEGER,
Spending_Score INTEGER)''')
print("TABLE CREATED SUCCESSFULLY.")
```

```python
for row in df.itertuples():
    cur.execute('''INSERT INTO SPENDING_ANALYSIS (CUSTOMERID, GENRE, AGE, Annual Income (k$), Spending_Score) VALUES(?,?,?,?,?)''', (row.CustomerID, row.Genre, row.Age, row.Annual Income (k$), row.Spending_Score))
conn.commit()
cur.close()
print("Records inserted successfully.")
```
